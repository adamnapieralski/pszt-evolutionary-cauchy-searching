```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_from_file(name, cauchy_file, normal_file, limit=None):
    cauchy = np.loadtxt(cauchy_file)
    normal = np.loadtxt(normal_file)
    if limit is not None:
        cauchy = cauchy[:limit]    
        normal = normal[:limit]    
        
    plt.plot(cauchy, label = 'cauchy')
    plt.plot(normal, label = 'normal')
    plt.title(name)
    plt.xlabel('Liczba iteracji')
    plt.ylabel('Błąd')
    plt.legend()
    fname = name.replace(' ', '_').lower()
    plt.savefig('analysis_out/figures/' + fname + '.pdf')
```

## 1 Funkcja Bent Cigar

```python
plot_from_file("Funkcja Bent Cigar 2D",
               'analysis_out/bent_cigar/bent_cigar_progress_cauchy__D2_P50_I1000_R5120200416_164942.txt', 
               'analysis_out/bent_cigar/bent_cigar_progress_normal__D2_P50_I1000_R5120200416_164942.txt')
```
