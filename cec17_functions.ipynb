```python
# %matplotlib notebook
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import cec17.functions as cec
from mpl_toolkits.mplot3d import Axes3D
```

```python
def plot_cec_function(function_name, k = 100, xy_range=(-50,50),
                      modification = True, random_modification = False):
    
    x = np.array(np.linspace(xy_range[0], xy_range[1], k))
    y = x
    xx, yy = np.meshgrid(x, y, indexing='ij')
    population = np.array([xx, yy]).transpose(1, 2, 0).reshape(k**2, 2)
    f = cec.modal_function(function_name, population, modification, random_modification)
   
    fig = plt.figure(figsize=(16, 8))
    fig.suptitle(function_name, fontsize=20)
    
    ax = fig.add_subplot(1, 2, 1, projection='3d')
    ax.plot_surface(xx, yy, f.reshape(k,k), cmap='viridis')
    ax.set_title('Surface plot')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')

    ax2 = fig.add_subplot(1, 2, 2, projection='3d')
    ax2.view_init(azim=-90, elev=90)
    ax2.plot_surface(xx, yy, f.reshape(k,k), cmap='viridis', edgecolor='none')
    ax2.zaxis.set_ticklabels([])
    ax2.set_xlabel('x')
    ax2.set_ylabel('y')
    plt.show()
```

```python
plot_cec_function('bent_cigar')
```

```python
plot_cec_function('zakharov')
```

```python
plot_cec_function('rosenbrock')
```

```python
plot_cec_function('rastrigin', xy_range=(-50, 50), modification = True, random_modification = False)
```
