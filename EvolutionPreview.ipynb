```python
s = np.random.normal(0, 1, 1000000)
plt.hist(s, bins=100)
plt.show()
```

```python
s = np.random.standard_cauchy(1000000)
s = s[(s>-4) & (s<4)]  # truncate distribution so it plots well
plt.hist(s, bins=100)
plt.show()
```

## EvolutionAlg class

```python
import numpy as np
from evolutionalg import *
```

```python
def fitness(population):
    return [np.abs((x[0] + x[1])) for x in population]
```

```python
e = EvolutionAlg()
population = np.array([[1,2], [3,4], [3,1], [4,2], [5,3], [6,2]])

e.run(population, fitness, 20, 4, mutation='normal', mutation_std=1)
```

```python
population = np.array([[1,2], [3,4], [3,1], [4,2], [5,3], [6,2]])
e.run(population, fitness, 20, 4, mutation='cauchy')
```
